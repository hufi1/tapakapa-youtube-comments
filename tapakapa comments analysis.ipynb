```python
import numpy as np  # numpy importieren
import pandas as pd  # pandas importieren
import matplotlib.pyplot as plt  # pyplot importieren
import seaborn as sns
```

```python
kommentare = pd.read_csv(r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\comments.csv') # datei einlesen
```

```python
kommentare.head() #kleiner check, was wir denn so im dataframe haben
```

```python
#check, ob leere kommentare vorhanden sind
kommentare['Comment Content'].isna().sum()
```

```python
from textblob import TextBlob #hier importieren wir textblob zur textanalyse.
```

```python
stimmung = []
for i in kommentare['Comment Content']:                  # hier checken wir für jede Zeile in der Spalte "Comment Content" die 
    einzelstimmung = TextBlob(i).sentiment.polarity      # Stimmung mittels TextBlob
    stimmung.append(einzelstimmung)                      # und fügen sie an die Liste "Stimmung" an.
```

```python
kommentare['Stimmung'] = stimmung                   # die Spalte Stimmung wird kreiert und mit der gleichnamigen Liste befüllt
kommentare.head()
```

```python
kommentare.sort_values(by='Stimmung', ascending=False)        #erstmal sortieren für mehr Überblick :)
```

```python
positive_kommentare = kommentare[kommentare['Stimmung']>0]        # wir kreieren 2 subsets, um diese dann auswerten zu können
negative_kommentare = kommentare[kommentare['Stimmung']<0]        
```

```python
positive_kommentare.sample(3)
```

```python
negative_kommentare.sample(3)
```

```python
from wordcloud import WordCloud, STOPWORDS # hier importieren wir die WordCloud und ebenso STOPWORDs, welches füllwörter enthält
```

```python
#um die wordcloud zu bilden, erstellen wir erstmal eine var, die alle wörter unserer comments enthält.
pos_kommentare_joined = ' '.join(positive_kommentare['Comment Content'])
neg_kommentare_joined = ' '.join(negative_kommentare['Comment Content'])
```

```python
# natürlich wollen wir, dass das alles auch schick aussieht :) 
# deshalb bestimmen wir eine eigene Font und auch eine Form (Mask)
font_path = r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\Grobold.ttf'
from PIL import Image
mask_pos = np.array(Image.open(r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\thumbs_up.jpg'))
mask_neg = np.array(Image.open(r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\thumbs_down.jpg'))
```

```python
#nun generieren wir die positive Wordcloud, ziehen aber die Füllwörter sowie einige generische Austria/Australia-Begriffe ab
stop_words = ['Austria', 'Australia', 'Österreich', 'Australien', 'Austrian', 'Australian', 'german', 'video', 'people', 'Europe'] + list(STOPWORDS)
wordcloud_pos = WordCloud(width=1600,
               height=800, stopwords=stop_words, background_color='white', font_path=font_path, mask=mask_pos).generate(pos_kommentare_joined)
plt.figure(figsize=(25,10))
plt.imshow(wordcloud_pos)
plt.axis('off')
```

```python
# und jetzt nochmal die negative Worcloud
stop_words = ['Austria', 'Australia', 'Österreich', 'Australien', 'Austrian', 'Australian', 'german', 'video', 'people', 'Europe'] + list(STOPWORDS)
wordcloud_neg = WordCloud(width=mask_neg.shape[1],
               height=mask_neg.shape[0], stopwords=stop_words, background_color='white', font_path=font_path, mask=mask_neg).generate(neg_kommentare_joined)
plt.figure(figsize=(25,10))
plt.imshow(wordcloud_neg)
plt.axis('off')
```

```python
# diese 2 zeilen exportieren die beiden clouds 
wordcloud_pos.to_file(r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\wordcloud_pos.jpg')
wordcloud_neg.to_file(r'C:\Users\Sabrina\Desktop\1-Youtube Text Data Analysis\wordcloud_neg.jpg')
```

```python
# wir sehen uns nun an, wann die meisten Kommentare abgegeben wurden & wie positiv diese ausfielen.
import datetime as dt
kommentare['Comment Published Date'] = pd.to_datetime(kommentare['Comment Published Date'])
```

```python
kommentare['Date'] = kommentare['Comment Published Date'].dt.date
```

```python
kommentare_nach_tag = kommentare.groupby('Date').agg({'Comment ID': 'count', 'Stimmung':'sum'}).reset_index()
kommentare_nach_tag = kommentare_nach_tag.rename(index=str, columns={'Comment ID' : 'Comments'})
```

```python
kommentare_nach_tag = kommentare_nach_tag.sort_values(by='Date', ascending=True)
kommentare_nach_tag
```

```python
# je dunkler der Rotton, desto positiver die Kommentare des jeweiligen Tages.
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(12, 6))
chart = sns.scatterplot(data= kommentare_nach_tag, x='Date', y='Comments', hue='Stimmung', palette = "Reds")
ax.set_xlim([dt.date(2018, 6, 25), dt.date(2020, 10, 3)])
ax.set_ylim(0, 50)
plt.show()
```

```python
# es föllt auf, dass jeweils im Oktober viele Kommentare abgegeben wurden, was auf besonders hohen Traffic zu dieser Jahreszeit
# schließen lässt. Ein Blick auf den Upload-Schedule verrät, dass es zumindest nicht direkt mit tapakapa's Uploadzeiten korreliert, 
# da der Youtuber über das ganze Jahr verteilt Videos hochlädt.
```

```python
# wir checken noch, wie viele kommentare die einzelnen User abgegeben haben.
kommentare_user = kommentare.groupby('Comment Author Name')['Comment ID'].count().reset_index()
kommentare_user = kommentare_user.rename(index=str, columns={'Comment ID':'Comments'})
kommentare_user = kommentare_user.sort_values(by='Comments', ascending=False)
```

```python
kommentare_user.head()
# scheinbar gibt es keine 'Super Poster'
```

```python
# wir checken weiters, ob es gewisse 'Spam-Kommentare' gibt. [= Kommentare die, häufig abgegeben werden, um den YT-Algorithmus zu beeinflussen
# bzw auch um eigenen Content oder Produkte zu bewerben.]
kommentare_nach_content = kommentare.groupby('Comment Content')['Comment ID'].count().reset_index()
kommentare_nach_content = kommentare_nach_content.rename(index=str, columns={'Comment ID': 'Comment_Amt'}).sort_values(by='Comment_Amt', ascending=False)
kommentare_nach_content.head()
# offenbar nicht, der häufigste Kommentar ist ebenso recht unscheinbar.
```
